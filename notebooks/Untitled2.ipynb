```python
# import usuals
from math import exp, sqrt, pi
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import seaborn as sns
import warnings

# some config
%config Completer.use_jedi = False
sns.set()
warnings.filterwarnings('ignore')
```

```python
def norm_pdf(x, sigma=1):
    
    return (1 / sqrt(2*pi)) * (1 / sigma) * exp(-0.5 * (x/sigma)**2)
```

```python
xs = np.linspace(-4, 4, 8*200 + 1)
```

```python
deltas = [0, 0.05, 0.1, 0.2, 0.5]
```

```python
perturbs = []

for d in deltas:
    p = [0.5 * (norm_pdf(x, 1 + d) + norm_pdf(x, 1 - d)) for x in xs]
    perturbs.append([xs, p])
```

```python
fig, ax = plt.subplots(figsize=(20,10))

for dist, d in zip(perturbs, deltas):
    ax.plot(dist[0], dist[1], label=d)

ax.legend();
```
