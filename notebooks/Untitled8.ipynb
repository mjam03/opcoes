```python
# usual suspects
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import warnings

# some config
%config Completer.use_jedi = False
warnings.filterwarnings('ignore')
sns.set()
```

```python
n = 10000
sums = [1, 2, 3, 4, 10]

rng = np.random.default_rng(seed=123)

uni_sums = {}
for s in sums:
    d = rng.uniform(size=(n, s)).sum(axis=1)
    uni_sums[s] = d

# plot them
fig, axes = plt.subplots(ncols=5, figsize=(32,8))
axes = axes.reshape(-1)

for s, ax in zip(sums, axes):
    sns.distplot(uni_sums[s], kde=False, norm_hist=True, fit=stats.norm, ax=ax)
    # label them
    ax.set_title("Summands: {}".format(s))
    ax.set_xlabel("x", fontsize=20)
    ax.set_ylabel("Density", fontsize=20)
    
fig.suptitle("Uniform Convergence to Normal", fontsize=24);
```

```python
stats.kstest(uni_sums[sums[-1]], "norm")
stats.kstest((uni_sums[sums[-1]] - np.mean(uni_sums[sums[-1]])) / np.std(uni_sums[sums[-1]]), "norm")
```

```python
stats.shapiro(uni_sums[sums[-1]])
```

```python
fig, ax = plt.subplots(ncols=1, figsize=(12, 12))
stats.probplot(uni_sums[sums[-1]], dist="norm", plot=ax);
```

```python
n = 10000
sums = [1, 3, 10, 30, 10000]

rng = np.random.default_rng(seed=123)

gamma_sums = {}
for s in sums:
    d = rng.gamma(1, scale=1/1, size=(n, s)).sum(axis=1)
    gamma_sums[s] = d

# plot them
fig, axes = plt.subplots(ncols=5, figsize=(32,8))
axes = axes.reshape(-1)

for s, ax in zip(sums, axes):
    sns.distplot(gamma_sums[s], kde=False, norm_hist=True, fit=stats.norm, ax=ax)
    # label them
    ax.set_title("Summands: {}".format(s))
    ax.set_xlabel("x", fontsize=20)
    ax.set_ylabel("Density", fontsize=20)
    
fig.suptitle("Gamma Convergence to Normal", fontsize=24);
```
